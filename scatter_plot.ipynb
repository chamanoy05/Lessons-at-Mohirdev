```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
plt.figure(figsize=(5, 7))
sns.scatterplot(x=df['price'], y=df['mileage'])
```

```python
plt.figure(figsize=(8, 9))
sns.scatterplot(x='year', y='price', data=df)
plt.show()
```

.regplot yordamida o'sish yoki kamayishni tasvirlash mumkin va bunday holatda to'g'ri chiziq orqali ko'rsatilinadi

```python
df= df[df['year']!=1970]# 1970 yildagi qiymatni olib tashlayappiz
sns.regplot(x='year', y='price', data=df)
```

```python
#grafikka uchunchi o'zgaruvchi qo'shish
sns.scatterplot(x='year', y='price', hue='transmission', data=df)
```

```python
#grafikka to'rtinchi o'zgaruvchini qo'shish
sns.scatterplot(data=df, x='year', y='price', hue='transmission', style='fuelType', s=100)#s bu yerda nuqtalar soni
plt.show()
```

scatterplotning yana bir ko'rinishi .swarmplot
swarmplot orqali yuqoridagi har bir kategoriyani alohida swarm ko'rinishida keltirish mumkin

```python
sns.swarmplot(x='year', y='price', data= df)
plt.show()#kod bajarilishi ucchun doimo ko'p vaqt oladi
```
